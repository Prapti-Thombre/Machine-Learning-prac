```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
#data cleaning
dataset[dataset.isnull().any(axis=1)]
```

the dataste is converted into two sets based on region from 122th index, can make a new column based on the region

1 : "BEjaia Region dataset"

2: " Sidi - Bel abbes Region dataset"

add a new column region

```python
df = dataset
df.loc[:122,"Region"] = 0
df.loc[122:, "Region"] = 1
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
##remove the null values
df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#remove the 122nd row which is having null value
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
#fix spaces in column names
df.columns = df.columns.str.strip()
df.columns
```

```python
##change the required columns as integer datatype
df[['month','day','year','Temperature','RH','Ws']] = df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
#changing the other columns to float datatype
objects = [features for features in df.columns if df[features].dtypes == 'o']
```

```python
for  i in objects:
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.head()
```

```python
#3lets save the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```
